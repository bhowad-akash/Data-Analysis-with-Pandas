### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users and use the 'user_id' as index

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|').set_index('user_id')
```

```python
users.head(25)
```

```python
users.tail(10)
```

```python
users.shape[0]
```

### Step 7. What is the number of columns in the dataset?

```python
users.shape[1]
```

### Step 8. Print the name of all the columns.

```python
users.columns
```

### Step 9. How is the dataset indexed?

```python
users.index
```

### Step 10. What is the data type of each column?

```python
users.dtypes
```

### Step 11. Print only the occupation column

```python
users['occupation']
```

### Step 12. How many different occupations are in this dataset?

```python
users['occupation'].value_counts().count()
```

### Step 13. What is the most frequent occupation?

```python
users['occupation'].value_counts().sort_values(ascending = False).head(1).index[0]
```

### Step 14. Summarize the DataFrame.

```python
users.describe()
```

### Step 15. Summarize all the columns

```python
users.describe(include='all')
```

### Step 16. Summarize only the occupation column

```python
users['occupation'].describe()
```

### Step 17. What is the mean age of users?

```python
users['age'].mean().round(2)
```

### Step 18. What is the age with least occurrence?

```python
users['age'].value_counts().sort_values().head(5).index[0:5]
```
